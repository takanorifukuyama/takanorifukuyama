```julia
# Jf2-26
using Plots

x = 0:0.01:50
plot(x, cos.(x), c="red", label="No Dumping")
plot!(x, exp.(-x/10).*cos.(x), c="blue", label="Dumping")
```

```julia
# Jf2-27 Morse Curve
using Plots

r = 0.2:0.05:2.85
D = 430.862
Re = 0.75
a = 1.94
morse(r) = @.D * (1 - exp(-a*(r-Re)))^2
Plots.plot(r, morse(r), marker=:circle, 
            title="Morse Curve of H2",
            xlabel="Bound Length [Angstroms]",
            ylabel="Energy [kJ/mol]",
            legend=false)
```
